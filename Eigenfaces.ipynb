```python
from sklearn import datasets
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

from sklearn import set_config
set_config(print_changed_only=False)
```

```python
oliv = datasets.fetch_olivetti_faces()
```

```python
fig = plt.figure(figsize=(8,8))
for  i in range(64):
    ax = fig.add_subplot(8, 8, i+1)
    ax.imshow(oliv.images[i], cmap=plt.cm.gray)
  

plt.show()
```

```python
x = oliv.data
y = oliv.target
```

```python
pca = PCA()
pca.fit(x)
```

```python
pca.components_.shape
```

```python
#Finding Optimal Number Of Features

k = 0 
total = sum(pca.explained_variance_)
currentSum = 0
while currentSum/total < 0.95:
    #currentSum += pca.explained_variance_[k]
    currentSum = currentSum + pca.explained_variance_[k]
    k = k + 1
    
k
```

```python
#Compressing the data by reducing the dimensionality using PCA
pca = PCA(n_components=k, whiten=True)
tranformed_data = pca.fit_transform(x)
tranformed_data.shape
```

### Reconstructing the data so that it returns to its original dimension by inverse the transformation

##### There will be an information losses, we cant reconstruct the original data 100%

```python
x_approx = pca.inverse_transform(tranformed_data)
x_approx.shape
```

```python
x_approx_images = x_approx.reshape((400, 64, 64))
```

```python
fig = plt.figure(figsize=(8,8))
for  i in range(64):
    ax = fig.add_subplot(8, 8, i+1)
    ax.imshow(x_approx_images[i], cmap=plt.cm.gray)
  

plt.show()
```
